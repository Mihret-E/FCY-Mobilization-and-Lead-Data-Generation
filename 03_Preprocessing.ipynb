```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

import joblib
```

```python
customers = pd.read_csv("customers.csv")

customers.head()
```

```python

customers.info()
```

```python
#Check Missing Values
customers.isnull().sum()
```

```python
#Remove Duplicate Records
customers = customers.drop_duplicates()
```

```python
customers.shape
```

```python
#Convert Yes/No Columns
binary_map = {
    "Yes":1,
    "No":0
}
```

```python
customers["FCYAccount"] = customers["FCYAccount"].map(binary_map)

customers["ETBAccount"] = customers["ETBAccount"].map(binary_map)

customers["MobileBanking"] = customers["MobileBanking"].map(binary_map)

customers["InternetBanking"] = customers["InternetBanking"].map(binary_map)
```

```python
customers.head()
```

```python
#Encode Gender
gender_encoder = LabelEncoder()
```

```python
customers["Gender"] = gender_encoder.fit_transform(
    customers["Gender"]
)
```

```python
#Encode Occupation
occupation_encoder = LabelEncoder()

customers["Occupation"] = occupation_encoder.fit_transform(
    customers["Occupation"]
)
```

```python
#Encode Region
region_encoder = LabelEncoder()

customers["Region"] = region_encoder.fit_transform(
    customers["Region"]
)
```

```python
#Encode Branch
branch_encoder = LabelEncoder()

customers["Branch"] = branch_encoder.fit_transform(
    customers["Branch"]
)
```

```python
customers.head()
```

```python
#Separate Features and Target
y = customers["LeadStatus"]
```

```python
X = customers.drop(
    columns=["LeadStatus"]
)
```

```python
print(X.shape)

print(y.shape)
```

```python
#Train/Test Split
X_train,X_test,y_train,y_test = train_test_split(

X,

y,

test_size=0.20,

random_state=42,

stratify=y

)
```

```python
print(X_train.shape)

print(X_test.shape)
```

```python
#Train/Test Split
numeric_columns = [

"Age",

"MonthlyIncome",

"RemittanceCount",

"TotalRemittanceUSD",

"AverageTransactionUSD",

"FCYPurchaseUSD",

"LastTransactionDays",

"ExistingProducts"

]
```

```python
scaler = StandardScaler()
```

```python
X_train[numeric_columns] = scaler.fit_transform(
    X_train[numeric_columns]
)
```

```python
X_test[numeric_columns] = scaler.transform(
    X_test[numeric_columns]
)
```

```python
#Verify Scaling
X_train.head()
```

```python
#Save Encoders
joblib.dump(
    gender_encoder,
    "gender_encoder.pkl"
)

joblib.dump(
    occupation_encoder,
    "occupation_encoder.pkl"
)

joblib.dump(
    region_encoder,
    "region_encoder.pkl"
)

joblib.dump(
    branch_encoder,
    "branch_encoder.pkl"
)
```

```python
#Save Scaler
joblib.dump(
    scaler,
    "scaler.pkl"
)
```

```python
#Save Processed Data
X_train.to_csv(
    "X_train.csv",
    index=False
)

y_train.to_csv(
    "y_train.csv",
    index=False
)
```

```python
X_test.to_csv(
    "X_test.csv",
    index=False
)

y_test.to_csv(
    "y_test.csv",
    index=False
)
```
