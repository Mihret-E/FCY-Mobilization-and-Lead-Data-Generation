```python
import pandas as pd
import numpy as np

np.random.seed(42)
```

```python
#Number of Customers

n = 500
customer_id = np.arange(100001, 100001 + n)
```

```python
#customer age
age = np.random.randint(21, 70, n)
```

```python
#gender
gender = np.random.choice(
    ['Male', 'Female'],
    size=n,
    p=[0.60,0.40]
)
```

```python
#occupation
occupation = np.random.choice(

[
    'Government Employee',
    'Private Employee',
    'Business Owner',
    'Merchant',
    'Farmer',
    'Engineer',
    'Doctor',
    'Teacher',
    'NGO Employee',
    'Student'
],

size=n,

p=[0.18,
   0.18,
   0.14,
   0.12,
   0.08,
   0.07,
   0.05,
   0.08,
   0.05,
   0.05]
)
```

```python
#monthly income
monthly_income = np.random.randint(
    12000,
    180000,
    n
)
```

```python
#region

region = np.random.choice(

[
    "Addis Ababa",
    "Oromia",
    "Amhara",
    "Tigray",
    "Sidama",
    "SNNPR",
    "Harari",
    "Dire Dawa"
],

size=n
)
```

```python
#branch
branch = np.random.choice(

[
    "Bole",
    "Mexico",
    "Piassa",
    "CMC",
    "Megenagna",
    "Adama",
    "Bahir Dar",
    "Hawassa"
],

size=n
)
```

```python
#Existing FCY Account

#Assume only about 30% already have one.
fcy_account = np.random.choice(

['Yes','No'],

size=n,

p=[0.30,0.70]
)
```

```python
#ETB Account

#Almost every customer has an ETB account.

etb_account = np.random.choice(

['Yes','No'],

size=n,

p=[0.95,0.05]
)
```

```python
#mobile banking
mobile_banking = np.random.choice(

['Yes','No'],

size=n,

p=[0.75,0.25]
)
```

```python
#internet banking
internet_banking = np.random.choice(

['Yes','No'],

size=n,

p=[0.45,0.55]
)
```

```python
#remmitance count
remittance_count = np.random.poisson(7,n)
```

```python
#total remmitance (usd)

total_remittance = remittance_count * np.random.randint(

300,

1200,

n

)
```

```python
#fcy purchase
#Customers normally purchase a portion of their remittance.

fcy_purchase = (

total_remittance *

np.random.uniform(

0.20,

0.95,

n

)

).astype(int)
```

```python
#last transaction day
last_transaction = np.random.randint(

1,

365,

n
)
```

```python
#existing banking product
existing_products = np.random.randint(

1,

6,

n
)
```

```python
#average transaction
average_transaction = (

total_remittance /

(remittance_count + 1)

).round(2)
```

Create the Target Variable (LeadStatus)

Since we don't have historical campaign data, we'll use a business rule to simulate which customers are likely to convert.

A customer is considered a lead if:

They do not already have an FCY account.
They received at least 8 remittances in the past year.
Their total remittance is at least USD 4,000.
They use Mobile Banking.

```python
lead_status = np.where(
    (
        (fcy_account == "No") &
        (remittance_count >= 8) &
        (total_remittance >= 4000) &
        (mobile_banking == "Yes")
    ),
    1,
    0
)
```

```python
#create data frame
customers = pd.DataFrame({

'CustomerID':customer_id,

'Age':age,

'Gender':gender,

'Occupation':occupation,

'MonthlyIncome':monthly_income,

'Region':region,

'Branch':branch,

'FCYAccount':fcy_account,

'ETBAccount':etb_account,

'MobileBanking':mobile_banking,

'InternetBanking':internet_banking,

'RemittanceCount':remittance_count,

'TotalRemittanceUSD':total_remittance,

'AverageTransactionUSD':average_transaction,

'FCYPurchaseUSD':fcy_purchase,

'LastTransactionDays':last_transaction,

'ExistingProducts':existing_products,

'LeadStatus':lead_status

})
```
